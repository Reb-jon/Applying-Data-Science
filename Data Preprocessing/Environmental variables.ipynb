## AQI Air Quality Index data
https://www.kaggle.com/datasets/adampq/air-quality-index-by-state-1980-2022

Impact: Poor air quality can be associated with higher health risks and stress levels, which might impact an individual's ability to maintain a stable income or financial stability.

This kaggle dataset takes measurements from the AQI website and records air quality data by state and year.




## Natural Disaster rate by state 2024
https://worldpopulationreview.com/state-rankings/natural-disasters-by-state

Impact: Areas that are prone to extreme weather events, such as hurricanes, floods, or droughts, may have more financial instability. This could impact an individual's ability to make timely payments or even disrupt their employment or business operations.

This dataset provides the natural disaster rate for all 50 states.



## State profile and energy estimates
https://www.eia.gov/state/compare/?sid=AL#?selected=US-AL-AK-AZ-AR-CA-CO-CT-DE-DC-FL-GA-HI-ID-IL-IN-IA-KS-KY-LA-ME-MD-MA-MI-MN-MS-MO-MT-NE-NV-NH-NJ-NM-NY-NC-ND-OH-OK-OR-PA-RI-SC-SD-TN-TX-UT-VT-VA-WA-WV-WI-WY

This dataset provides data on energy production, consumption, and efficiency by state. For example, states with higher reliance on fossil fuels versus renewables could have different economic challenges.  It also includes data related to energy sources, emissions, and energy efficiency, which could be used to identify environmental risks that may correlate with economic hardship and, consequently, credit risk.
